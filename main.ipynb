```python
import os
import requests
from IPython.display import Markdown, display
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)

openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
        print(f"Keys are set up and they start with {openai_api_key[:10]}")
```

```python
system_prompt = """
You are a professional Personal AI Assistant with access to many tools, that you can pick from to optimize the workflow.
Ur goal is to be helpful and easy to communicate with.
"""
```

```python
user_prompt = """
Hello, what can you do for me?
"""
```

```python
messages = [
    {'role': 'system', 'content': system_prompt},
    {'role': 'user', 'content': user_prompt}
]

messages
```

```python
openai = OpenAI()

MODEL = 'gpt-5-nano'

response = openai.chat.completions.create(model=MODEL, messages=messages)
display(Markdown(response.choices[0].message.content))
```

```python
tasks = []
```

```python
def add_task(title):
    tasks.append(title)
    return f"Task '{title}' added successfully."
```

```python
tools = [
    {
        "type": "function",
        "function": {
            "name": "add_task",
            "description": "Add a new task to the task list",
            "parameters": {
                "type": "object",
                "properties": {
                    "title": {
                        "type": "string"
                    }
                },
                "required": ["title"]
            }
        }
    }
]
```

```python
response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)
```

```python
display(Markdown(response.choices[0].message.content))
```

```python
user_prompt = """
Hello, can you set up a reminder task because i have to finish backpropagation at my project VHgrad.
"""
```

```python
response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)
display(Markdown(response.choices[0].message.content))
```

```python
import json
```

```python
message = response.choices[0].message

if message.tool_calls:
    tool_call = message.tool_calls[0]
    function_name = tool_call.function.name
    arguments = json.loads(tool_call.function.arguments)

    if function_name == "add_task":
        result = add_task(**arguments)

    messages.append(message)
    messages.append({
        "role": "tool",
        "tool_call_id": tool_call.id,
        "content": result
    })

    final_response = openai.chat.completions.create(
        model=MODEL,
        messages=messages
    )

    print(final_response.choices[0].message.content)

else:
    print(message.content)
```
