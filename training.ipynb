```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
import gradio as gr
```

```python
MODEL = 'gpt-5-nano'
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"Successful, current key starts with {openai_api_key[:15]}")
```

```python
message = [{'role':'user', 'content':'My name is virvi huta'}]

message
```

```python
openai = OpenAI()

response = openai.chat.completions.create(model='gpt-5-mini', messages=message)
response.choices[0].message.content
```

```python
def message_gpt(prompt):
    message = [{'role': 'user', 'content': prompt}]
    response = openai.chat.completions.create(model=MODEL, messages=message)
    return response.choices[0].message.content
```

```python
def stream_gpt(prompt):
    message = [{'role':'user', 'content':prompt}]
    stream = openai.chat.completions.create(model=MODEL, messages=message, stream=True)
    result = ''
    for chunk in stream:
        result += chunk.choices[0].delta.content or ''
        yield result
```

```python
def ask_for_input():
    answer = input("Hello, want to talk to gpt?")
    if answer == 'Yes' or answer == 'yes':
        type = input('Do you want GPT to stream or to reply immediately?')
        if type == 'stream' or type == 'Stream':
            print('Streaming it is: ')
            prompt = input('What do you want to say to GPT? ')
            for chunk in stream_gpt(prompt):
                print(chunk)
        elif type == 'message' or type == 'Message':
            prompt = input('What do you want to say to GPT? ')
            reply = message_gpt(prompt)
            print(reply)
        else:
            print('Unknown answer')
    else:
        print('Okay bye')
```
